```python
def create_table(name, selected_y):
    selected_y = f'{selected_y}'
    file = pd.read_csv(f'D:\Earth.Org\Low Temperatures\country_mod\{name}.csv')
    #print(file)
    file = file[file.columns[1:]]
    #print(file)
    years = file.columns[2:]
    if selected_y in years:
        #print('found')
        pass
    elif name == 'Greenland':
        selected_y = '2015'
    else:
        selected_y = years[-1]
    index = 0
    while years[index] != selected_y:
        index += 1
    years = years[:index]
    #print(years)
    mean_temp = file[file.columns[1]]
    file['referenceTemp(till 2000)'] = mean_temp
    #print(mean_temp)
    for y in years:
        mean_temp += file[y]
    mean_temp = mean_temp/(len(years)+1)
    file['meanTemp'] = mean_temp
    file['Country'] = name
    file['Deviation'] = file[selected_y] - file['meanTemp']
    file = file[['Country','Month', 'meanTemp', selected_y, 'Deviation']]
    # file = file.set_index('Country','Month')
    return file
```

```python
file = create_table('Austria', 2020)
file
```

```python
file.plot(title='Austria')
```

```python
data = pd.read_csv('D:\Earth.Org\Low Temperatures\coldest_regions.csv')
data = list(set(data.Country))
data
```

```python
file = None
for c in data:
    if file is None:
        file = create_table(c,2020)
    else:
        file = pd.concat([file,create_table(c,2020)])
file
```
