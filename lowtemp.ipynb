```python
import pandas as pd
import numpy as np
import geopandas as gpd
```

```python
df = pd.read_csv('GlobalLandTemperatures_GlobalLandTemperaturesByCountry.csv')
df
```

```python
df = df.dropna()
df = df.sort_values(by=['Country','dt'], ascending=True)
df = df.reset_index(drop=True)
df
```

```python
countries = list(set(df['Country']))
countries.sort()
countries
```

```python
world = pd.read_csv('world-regions-according-to-maddison.csv')
world = world.rename(columns={'Entity': 'Country'})
world
```

```python
df = pd.merge(df,world)
df
```

```python
df = df[['Country', 'dt', 'AverageTemperature', 'AverageTemperatureUncertainty', 'World regions according to Maddison (Maddison (2010))']]
df
```

```python
df_mod = df[df['dt'] < '2000-01-01']
df_mod
```

```python
df_mod2 = pd.DataFrame(columns=['Country', 'AverageTemperature', 'Region'])
for c in countries:
    temp = {}
    region = None
    for x in df_mod[df_mod['Country'] == c].values:
        if x[2] in temp:
            temp[x[2]] +=1
        else:
            temp[x[2]] = 1
        if region == None:
            region = x[-1]
    sum = 0
    tot = 0
    index = 0
    for x in temp:
        sum += x*temp[x]
        tot += temp[x]
    if tot != 0:
        df2 = {'Country': c, 'AverageTemperature':sum/tot, 'Region':region}
    else:
        df2 = {'Country': c, 'AverageTemperature':sum, 'Region':region}
    df_mod2 = df_mod2.append(df2,ignore_index = True)
df_mod2 = df_mod2.sort_values('Country')
df_mod2 = df_mod2.dropna()
df_mod2 = df_mod2.reset_index(drop = True)
```

```python
coldest_regions = df_mod2.sort_values(by=['AverageTemperature', 'Region'])
coldest_regions = coldest_regions.reset_index(drop = True)
coldest_regions = coldest_regions[:20]
coldest_regions.to_csv('coldest_regions.csv')
```
