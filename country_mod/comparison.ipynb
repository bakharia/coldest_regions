```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def create_table(name, selected_y):
    selected_y = f'{selected_y}'
    file = pd.read_csv(f'D:\Earth.Org\Low Temperatures\country_mod\{name}.csv')
    file = file[file.columns[1:]]
    years = file.columns[2:]
    if selected_y in years:
        #print('found')
        pass
    else:
        selected_y = years[-1]
    index = 0
    while years[index] != selected_y:
        index += 1
    years = years[:index]
    print(years)
    mean_temp = file[file.columns[1]]
    for y in years:
        mean_temp += file[y]
    mean_temp = mean_temp/(len(years)+1)
    file['meanTemp'] = mean_temp
    file = file[['Month','meanTemp', selected_y]]
    file = file.set_index('Month')
    return file
```

```python
China = create_table('China', 2020)
China
```

```python
China.plot(title='China')
```
